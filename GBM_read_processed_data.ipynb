```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

```python
folders = glob.glob('Bursts/*')

# format
# `lc_{burst}_{det}_{t90_start}_{t90}_{SNR}.txt `
## units sec,counts/sec
#  `spec_{burst}_{det}_{t90_start}_{t90}_{SNR}.txt `
## units keV,counts/sec/keV


for i in [0]:
    if os.path.isfile(folders[i]+'/done.txt'):
        
        files_lc = glob.glob(folders[i]+'/lc*.txt')
        plt.figure()
        for lc in files_lc:
            t,counts = np.loadtxt(lc)
            det = lc[34:36]
            plt.plot(t,counts,label=f'{det}')
            plt.ylabel('Rate [counts/s]')
            plt.xlabel('Time [s]')
            plt.legend(loc='best')
            
        files_spec = glob.glob(folders[i]+'/spec*.txt')
        plt.figure()
        for spec in files_spec:
            E,flux = np.loadtxt(spec)
            det = spec[36:38]
            plt.plot(E,flux,label=f'{det}')
            plt.yscale('log')
            plt.ylabel('Rate [counts/s/keV]')
            plt.xlabel('Energy [keV]')
            plt.legend(loc='best')

            
    else:
        continue
```
